```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

#Function to scaling data using standardization

def standard_scaler(X):
    #transpose the matrix X
    X_trans = X.T
    #Numbef of features
    n = X.shape[1]
    for i in range(n):
        mean = np.mean(X_trans[i, :])
        std = np.std(X_trans[i, :])
        print(mean, std)
        X_trans[i, :] = np.array([ (x-mean)/std for x in X_trans[i, :]])
    return X_trans.T

#Function to compute loss function

def compute_cost_function(X, y, theta):
    #Number of training sample
    m = X.shape[0]
    tmp = np.sum((X.dot(theta.T)-y)**2)
    return 1/(2*m)*tmp

#Batch gradient descent implementation

def batch_gradient_descent(X, y, learning_rate, epochs):
    #Number of traning samples
    m = X.shape[0]
    #Number of features
    n = X.shape[1]
    #Init coefficient
    theta = np.zeros(n).reshape(1, -1)
    #Compute initial value of loss function
    J = compute_cost_function(X, y, theta)
    print('Initial loss function :',J)
    #utility variables
    iter = 0
    min_J = [J]
    previous_J = J
    converged = False
    while not converged:
        gradient = np.array([ np.sum((X.dot(theta.T)-y)*(X[:, j].reshape(-1, 1))) for j in range(n)])
        theta = theta - 1/m*learning_rate*gradient
        J = compute_cost_function(X, y, theta)
        print('J :',J)
        if J >= previous_J or iter == epochs:
            converged = True
            print('End of learning !')
        min_J.append(J)
        previous_J = J
        iter += 1
    plt.plot(min_J, label='Min cost function')
    plt.legend()
    plt.show()
    return theta 

                             
if __name__=='__main__':
    #Load data from directory
    data = pd.read_csv('C:/CNTT/DATA/ex1data2.txt')
    #Features and value
    X = np.array(data.iloc[:,:2]).reshape(-1, 2).astype('float64')
    tmp_X = np.copy(X)
    y = np.array(data.iloc[:,2]).reshape(-1, 1).astype('float64')
    X = standard_scaler(X)
    #You can using Scaler from sklearn to normalize data before apply gradient descent
    #scaler = StandardScaler()
    #X = scaler.fit_transform(X)
    print(X)
    for i in range(X.shape[1]):
        print(np.mean(X[:, i]))
        print(np.std(X[:, i]))
    X0 = np.ones(X.shape[0]).reshape(-1, 1)
    X = np.column_stack((X0, X))
    theta = batch_gradient_descent(X, y, 0.01, 1000)
    print(theta)
    
    plt.scatter(tmp_X[:, 0], y, label='Area (m^2)')
    predict_y  = X.dot(theta.T)
    plt.legend()
    plt.show()
    for i in range(y.shape[0]):
        if predict_y[i] <= y[i]:
            print(predict_y[i]/y[i]*100)
        else:
            tmp = predict_y[i]-y[i]
            print(100 - tmp/y[i]*100)
```
